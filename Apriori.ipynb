```python
records=[]
for i in range(0,21):
  records.append([str(df.values[i,j]) for j in range (0,6)])
```

```python
association_rule_1 = apriori(records, min_support = 0.50, min_confidence = 0.7, min_lift = 1.2, min_length = 2)
association_results = list(association_rule_1)
```

```python
print(len(association_results))
```

```python
print(association_results)
```

```python
assn_rule_2=apriori(records,min_support=0.40,min_confidence=0.8,min_lift=1.2,min_length=2)
assn_results=list(assn_rule_2)
```
