```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("C:/Users/hridi/Desktop/Entrance/Work/Python/seeds-less-rows.csv")
```

```python
plt.scatter(df.area,df['grain_variety'])
plt.xlabel('area')
plt.ylabel('grain_variety')
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['area','groove_length']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head(10)
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.area,df1['groove_length'],color='green')
plt.scatter(df2.area,df2['groove_length'],color='red')
plt.scatter(df3.area,df3['groove_length'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('area')
plt.ylabel('groove_length')
plt.legend()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['groove_length']])
df['groove_length'] = scaler.transform(df[['groove_length']])
scaler.fit(df[['area']])
df['area'] = scaler.transform(df[['area']])
```

```python
df.head()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['area','groove_length']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
