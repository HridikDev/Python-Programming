```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
column_names = [
'SampleCode', 'ClumpThickness', 'CellSizeUniformity', 'CellShapeUniformity',
'MarginalAdhesion', 'SingleEpiCellSize', 'BareNuclei', 'BlandChromatin',
'NormalNucleoli', 'Mitoses', 'Class'
]
bc_df = pd.read_csv(
"C:/Users/hridi/Desktop/Entrance/Work/Python/breast-cancer-wisconsin.data",
names=column_names,
na_values='?',
dtype={'BareNuclei': 'float64'}
)
bc_df.head()
```

```python
bc_df.isnull().sum()
```

```python
print(bc_df.info())
print(bc_df.describe())
```

```python
bc_df['Class'] = bc_df['Class'].map({2: 'Benign', 4: 'Malignant'})
bc_clean = bc_df.copy()
bc_clean['BareNuclei'].fillna(bc_clean['BareNuclei'].median(), inplace=True)
print(bc_clean.isnull().sum())
```

```python
sns.countplot(x="Class", data=bc_clean)
plt.title('Diagnosis Class Distribution')
plt.show()
```

```python
bc_df.describe()
```

```python
sns.countplot(x="Class", hue="ClumpThickness", data=bc_clean)
plt.title('Diagnosis vs Clump Thickness')
plt.legend(title="Clump Thickness", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
sns.countplot(x="Class", hue="CellSizeUniformity", data=bc_clean)
plt.title('Diagnosis vs Uniformity of Cell Size')
plt.legend(title="Cell Size Uniformity", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
sns.histplot(bc_clean["ClumpThickness"], bins=10, kde=True)
plt.title('Distribution of Clump Thickness')
plt.show()
```
