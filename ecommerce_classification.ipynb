```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, roc_auc_score, confusion_matrix, roc_curve, classification_report)
from sklearn.impute import SimpleImputer
from collections import Counter

# For imbalanced data (optional)
# pip install imbalanced-learn
from imblearn.over_sampling import SMOTE
```

```python
# If you have CSV (Kaggle variant), update filename accordingly:
df = pd.read_csv('data.csv', encoding='ISO-8859-1')  # adjust filename
```

```python
# 2) Basic exploration
print("Shape:", df.shape)
display(df.head())
display(df.info())
display(df.describe(include='all'))
```

```python
# show number of nulls
print("Missing per column:\n", df.isna().sum())
```

```python
# 3) Create target label: is_return = 1 if InvoiceNo starts with 'C'
df['is_return'] = df['InvoiceNo'].astype(str).str.startswith('C').astype(int)
```

```python
# 4) Basic cleaning
# Drop duplicates
before = df.shape[0]
df = df.drop_duplicates()
print(f"Dropped {before - df.shape[0]} duplicate rows")
```

```python
# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
```

```python
# Remove rows with null InvoiceDate (if any)
df = df[df['InvoiceDate'].notna()]
```

```python
# Replace missing descriptions (if any) with 'Unknown'
df['Description'] = df['Description'].fillna('Unknown')
```

```python
# Compute total price
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
# Remove negative or zero quantities (these often indicate returns or data issues)
# But since returns are labeled via InvoiceNo starting with 'C' we can keep negative quantities 
# if you want to treat them as returns. For the classification target we rely on is_return.
# For modeling, remove rows with Quantity==0 to avoid noise:
df = df[df['Quantity'] != 0]
```

```python
# 5) Quick correlations and aggregate stats
numeric_cols = ['Quantity', 'UnitPrice', 'TotalPrice']
display(df[numeric_cols].describe())
```

```python
# Correlation matrix
corr = df[numeric_cols + ['is_return']].corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation matrix')
plt.show()
```

```python
# 6) Feature engineering
# Date / time features
df['Month'] = df['InvoiceDate'].dt.month
df['DayOfWeek'] = df['InvoiceDate'].dt.dayofweek  # Monday=0
df['Hour'] = df['InvoiceDate'].dt.hour
```

```python
# Reduce Country categories: keep top 10 countries, rest as 'Other'
top_countries = df['Country'].value_counts().nlargest(10).index.tolist()
df['CountryReduced'] = df['Country'].where(df['Country'].isin(top_countries), 'Other')
```

```python
# Optionally group StockCode or Description by top frequent products
top_products = df['StockCode'].value_counts().nlargest(20).index.tolist()
df['TopProductFlag'] = df['StockCode'].isin(top_products).astype(int)
```

```python
# 7) Prepare modeling dataset
features = ['Quantity', 'UnitPrice', 'TotalPrice', 'Month', 'DayOfWeek', 'Hour', 'TopProductFlag', 'CountryReduced']
X = df[features].copy()
y = df['is_return']
```

```python
# One-hot encode CountryReduced
X = pd.get_dummies(X, columns=['CountryReduced'], drop_first=True)
```

```python
# 8) Train/test split (stratify because returns are comparatively rare)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print("Train label distribution:", Counter(y_train))
print("Test  label distribution:", Counter(y_test))
```

```python
# 9) Scaling numeric features
numeric_feats = ['Quantity', 'UnitPrice', 'TotalPrice', 'Month', 'DayOfWeek', 'Hour']
scaler = StandardScaler()
X_train[numeric_feats] = scaler.fit_transform(X_train[numeric_feats])
X_test[numeric_feats] = scaler.transform(X_test[numeric_feats])
```

```python
# 10) Optionally handle imbalance via SMOTE (only apply to training)
sm = SMOTE(random_state=42)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)
print("After SMOTE:", Counter(y_train_sm))
```

```python
# 11) Model 1: Logistic Regression
lr = LogisticRegression(max_iter=1000, class_weight='balanced')  # balanced helps with imbalance
lr.fit(X_train_sm, y_train_sm)
y_pred_lr = lr.predict(X_test)
y_proba_lr = lr.predict_proba(X_test)[:,1]
```

```python
# 12) Model 2: Random Forest
rf = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42, class_weight='balanced')
rf.fit(X_train_sm, y_train_sm)
y_pred_rf = rf.predict(X_test)
y_proba_rf = rf.predict_proba(X_test)[:,1]
```

```python
# 13) Evaluation metrics helper
def evaluate_model(y_true, y_pred, y_proba, name):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    roc = roc_auc_score(y_true, y_proba)
    print(f"\n{name} metrics:")
    print(f"Accuracy: {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall: {rec:.4f}")
    print(f"F1-score: {f1:.4f}")
    print(f"ROC AUC: {roc:.4f}")
    print("\nClassification report:\n", classification_report(y_true, y_pred, zero_division=0))
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d')
    plt.title(f'Confusion matrix ({name})')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
evaluate_model(y_test, y_pred_lr, y_proba_lr, "Logistic Regression")
evaluate_model(y_test, y_pred_rf, y_proba_rf, "Random Forest")
```

```python
# 14) ROC curves comparison
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_proba_lr)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_proba_rf)
plt.figure(figsize=(8,6))
plt.plot(fpr_lr, tpr_lr, label=f'Logistic (AUC={roc_auc_score(y_test, y_proba_lr):.3f})')
plt.plot(fpr_rf, tpr_rf, label=f'RandomForest (AUC={roc_auc_score(y_test, y_proba_rf):.3f})')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend()
plt.show()
```

```python
# 15) Feature importance (for RF)
importances = pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending=False)
display(importances.head(20))
plt.figure(figsize=(8,6))
sns.barplot(x=importances.head(20).values, y=importances.head(20).index)
plt.title("Top 20 Feature importances (Random Forest)")
plt.show()
```
