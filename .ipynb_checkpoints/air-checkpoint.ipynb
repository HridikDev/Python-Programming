```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler

# 1. Load Titanic Dataset
air = pd.read_csv("C:/Users/hridi/Desktop/Entrance/Work/Python/AirQualityUCI.csv", sep=';',decimal=",")
```

```python
air
```

```python
air["Datetime"] = pd.to_datetime(
air["Date"] + " " + air["Time"],
format="%d/%m/%Y %H.%M.%S", # Explicit format
errors="coerce"
)
# Drop old Date, Time columns
air = air.drop(columns=["Date", "Time"])
# Reorder columns (Datetime first)
air = air[["Datetime"] + [col for col in air.columns if col != "Datetime"]]
air.replace(-200, np.nan, inplace=True)
# Fill missing values (forward fill)
air.ffill(inplace=True)
print("Shape:", air.shape)
print("\nMissing values:\n", air.isna().sum())
print("\nData types:\n", air.dtypes)
```

```python
plt.figure(figsize=(12,6))
plt.plot(air["Datetime"],air["CO(GT)"],color="blue")
plt.title("CO Concentration Over Time")
plt.xlabel("Datetime")
plt.ylabel("CO (mg/m^3)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(air["C6H6(GT)"].dropna(), bins=50, kde=True, color="green")
plt.title("Distribution of Benzene (C6H6)")
plt.xlabel("C6H6 (Î¼g/m^3)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(air.corr(numeric_only=True), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Pollutants")
plt.show()
```

```python
daily_co = air.resample("D", on="Datetime")["CO(GT)"].mean()
plt.figure(figsize=(12,6))
plt.plot(daily_co.index, daily_co, color="red")
plt.title("Daily Average CO Concentration")
plt.xlabel("Date")
plt.ylabel("CO (mg/m^3)")
plt.show()
```
