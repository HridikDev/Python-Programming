```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Column names as per the Adult dataset description
col_names = [
"age", "workclass", "fnlwgt", "education", "education_num",
"marital_status", "occupation", "relationship", "race", "sex",
"capital_gain", "capital_loss", "hours_per_week", "native_country", "income"
]
# Read the data
adult_test = pd.read_csv(
'adult.test',
names=col_names,
skiprows=1, # skip the first line (if it's not actual data)
na_values='?',
sep=',',
skipinitialspace=True
)
adult_test.head()
```

```python
adult_test.isnull().sum()
```

```python
print(adult_test.info())
print(adult_test.describe(include='all'))
```

```python
# Remove rows with missing values (can also impute if desired)
adult_clean = adult_test.dropna().copy()
# Fix the income label which has trailing dot in 'adult.test'
adult_clean['income'] = adult_clean['income'].str.strip().str.replace('.', '', regex=False)
# Verify missing values are gone
print(adult_clean.isnull().sum())
```

```python
sns.countplot(x="income",data=adult_clean)
plt.title('Income Class Distribution (Test Set)')
plt.show()
```

```python
sns.countplot(x="income",hue="sex",data=adult_clean)
plt.title('Income vs Sex')
plt.show()
```

```python
sns.countplot(x="income",hue="education",data=adult_clean,order=['<=50K','>50K'],hue_order=adult_clean['education'].value_counts().index[:6])
# Top 6 education levels
plt.title('Income vs Education (Top 6)')
plt.legend(title="Education",bbox_to_anchor=(1.05,1),loc='upper left')
plt.show()
```

```python
sns.histplot(adult_clean["age"], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```
